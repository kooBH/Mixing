```python
import numpy as np
import matplotlib.pyplot as plt   
```

```python
N = 8
r = 0.1

def convert(r,angle) :
    theta= np.deg2rad(angle)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    print("[{:+.4f}, {:+.4f}, {:+.4f}],".format(x,y,0))
    return x,y

pt_x = []
pt_y = []
for i in range(N) : 
    x,y = convert(r,i*45)
    pt_x.append(x)
    pt_y.append(y)

fig,ax = plt.subplots(figsize=(6, 6))

circle = plt.Circle((0.0, 0.0), 0.1, edgecolor='black', facecolor='gray')
ax.add_artist(circle)

ax.scatter(pt_x,pt_y)

ax.set_aspect('equal')
ax.set_xlim(-0.15,0.15)
ax.set_ylim(-0.15,0.15)

plt.show()
```
