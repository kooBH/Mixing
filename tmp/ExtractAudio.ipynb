```python
import torch
import os,glob
from moviepy.editor import *
import numpy as np
import IPython.display as ipd
```

```python
path_sample = "../sample/00002.mp4"
```

```python
video = VideoFileClip(path_sample)
audio = video.audio
# raw = audio.to_soundarray()[:,0]

# Extract the audio as a list of samples
audio_samples = list(audio.iter_frames())
# Convert the list of samples to a NumPy array
raw = np.array(audio_samples)

print(raw.max())
print(raw.min())
print(raw.shape)

ipd.display(ipd.Markdown("## {}".format("raw")))
ipd.display(ipd.Audio(raw[:,0], rate=48000))
```
